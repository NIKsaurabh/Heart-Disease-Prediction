```python
dummy_dataset = pd.get_dummies(dataset, columns = ['sex', 'cp' , 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal'])
dummy_dataset.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
dummy_dataset[['age','trestbps', 'chol', 'thalach', 'oldpeak']] = sc.fit_transform(dummy_dataset[['age','trestbps', 'chol', 'thalach', 'oldpeak']])
dummy_dataset.head()
```

```python
X=dummy_dataset.drop(['target'], axis=1)
y=dummy_dataset['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
classifier = RandomForestClassifier(n_estimators = 50, random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn import metrics
accuracy = metrics.accuracy_score(y_test, y_pred)
print(accuracy)
```

```python
from sklearn.model_selection import cross_val_score
acc = cross_val_score(estimator = classifier, X=X_train , y=y_train, cv=10)
acc
```
